**Scenario 3: Relationship Between Variables**

Dataset Example: Advertising spend vs sales revenue.

Goal: Check correlation (does higher spend lead to more sales?).

Best Chart: Scatter Plot

Visualizes relationships between two continuous variables.

Helps detect correlation or outliers.

```python
np.random.seed(42)
spend = np.random.randint(1000, 5000, 50)
sales = spend * 0.8 + np.random.randint(100, 1000, 50)

plt.scatter(spend, sales)
plt.title("Advertising Spend vs Sales Revenue")
plt.xlabel("Ad Spend ($)")
plt.ylabel("Sales Revenue ($)")
plt.show()
```

**Word Clouds (Text Data Visualisation)**

Example 1 – Simple Word Cloud with Stopwords

Dataset: A short collection of customer reviews.

Teaching focus: Show how common words dominate and how removing stopwords changes results.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

text = """
This product is great, I love it! The quality is amazing and the price is fair.slow.slow
The service and quality was excellet nd quality!.slow.slow.slow. slow.slow.slow. slow.slow.slow
"""

wc1 = WordCloud(width=600, height=400, background_color="white").generate(text)

plt.imshow(wc1, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud Without Stopwords")
plt.show()

stop_words = set(STOPWORDS)
wc2 = WordCloud(width=600, height=400, background_color="white", stopwords=stop_words).generate(text)

plt.imshow(wc2, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud With Stopwords Removed")
plt.show()
```

**Word Cloud with Shape and Colors**

Dataset: A Python-related text block.

Teaching focus: Shape masking and color effects.

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image

python_text = "Python programming language AI machine learning data science Python coding analysis"

mask = np.array(Image.open("circle.jpg"))  # needs a white-background image

wc = WordCloud(width=600, height=400, background_color="white", mask=mask, contour_color="blue", contour_width=2,
               colormap="coolwarm").generate(python_text)

plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud in Circular Shape with Colors")
plt.show()
```

**Gantt Charts (Project Management Visualisation)**

Example 1 – Simple Project Timeline (Plotly)

```python
import plotly.express as px
import pandas as pd

tasks = [
    dict(Task="Planning", Start='2025-10-01', Finish='2025-10-05'),
    dict(Task="Data Collection", Start='2025-10-06', Finish='2025-10-10'),
    dict(Task="Data Cleaning", Start='2025-10-11', Finish='2025-10-15'),
    dict(Task="Visualization", Start='2025-10-16', Finish='2025-10-20'),
    dict(Task="Report Writing", Start='2025-10-01', Finish='2025-10-25')
]

df = pd.DataFrame(tasks)
fig = px.timeline(df, x_start="Start", x_end="Finish", y="Task", title="Project Gantt Chart")
fig.update_yaxes(autorange="reversed")  # Show tasks top-down
fig.show()
```

**Gantt Chart with Team Members and Progress**

Teaching focus: Adding responsible persons and completion percentages.

```python
tasks = [
    dict(Task="Planning", Start='2025-10-01', Finish='2025-10-05', Person="Alice", Completion=100),
    dict(Task="Data Collection", Start='2025-10-06', Finish='2025-10-10', Person="Bob", Completion=80),
    dict(Task="Data Cleaning", Start='2025-10-11', Finish='2025-10-15', Person="Charlie", Completion=50),
    dict(Task="Visualization", Start='2025-10-16', Finish='2025-10-20', Person="Dana", Completion=30),
    dict(Task="Report Writing", Start='2025-10-21', Finish='2025-10-25', Person="Eve", Completion=0)
]

df = pd.DataFrame(tasks)
fig = px.timeline(df, x_start="Start", x_end="Finish", y="Task", color="Person", title="Gantt Chart with Team & Progress")
fig.update_yaxes(autorange="reversed")
fig.show()
```
